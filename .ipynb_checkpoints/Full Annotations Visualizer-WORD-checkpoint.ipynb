```python
def parse_response(id_document='', language='en'):
    """ 
    parse QWAM responses
    :path: to response file
    """
    sub_folder = 'english_extract_' if language == 'en' else 'french_extract_'
    folder = f'data/cacib_jsonfull_20200218/{language}/{sub_folder}{docID[0:8]}/'
    response_file = folder + sub_folder + docID + '.json'
    with open(response_file) as json_file:
        data = json.load(json_file)
    return data if type(data) == dict else None
```

```python
language = 'en'
docID = "ATELBZ0020190916ef9g00004"
response = parse_response(id_document=docID, language=language)
pprint(response)
```

```python
# Parse entities, relations and return a list with same format
def get_qwam_annotations(response=None, entity_list=[]) -> list: 
    annotations = []
    
    # appending entities
    for entity in entity_list: 
        if entity in response['annotations'].keys():
            for _ in response['annotations'][entity]:
                annotations.append({
                    'type': 'Entity',
                    '_subtype': entity,
                    'string': _['string'],
                    'start': _['start'],
                    'end': _['end']
                })
    
    if 'Relation' in response['annotations'].keys():    
        # appending relations
        for rel in response['annotations']['Relation']: 
            annotations.append({
                'type': 'Relation',
                '_subtype': rel['type'],
                'string': rel['string'],
                'start': rel['start'],
                'end': rel['end']
            })
        
    return annotations
```

```python
entity_list = ['Company','Person','Location','City','Product', 'Organization']
annotations = get_qwam_annotations(response, entity_list)
annotations[0:5]
```
