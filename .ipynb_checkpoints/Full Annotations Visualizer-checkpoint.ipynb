```python
def parse_response(id_document='', language='en'):
    """ 
    parse QWAM responses
    :path: to response file
    """
    sub_folder = 'english_extract_' if language == 'en' else 'french_extract_'
    folder = f'data/cacib_jsonfull_20200218/{language}/{sub_folder}{docID[0:8]}/'
    response_file = folder + sub_folder + docID + '.json'
    with open(response_file) as json_file:
        data = json.load(json_file)
    return data if type(data) == dict else None
```

```python
language = 'en'
docID = "DATMON0020191029efao005lq"
response = parse_response(id_document=docID, language=language)
pprint(response)
```

```python
# Parse entities, relations and return a list with same format
def get_annotations(response=None, entity_list=[]) -> list: 
    annotations = []
    
    # appending entities
    for entity in entity_list: 
        if entity in response['annotations'].keys():
            for _ in response['annotations'][entity]:
                annotations.append({
                    'type': 'Entity',
                    '_subtype': entity,
                    'string': _['string'],
                    'start': _['start'],
                    'end': _['end']
                })
    
    if 'Relation' in response['annotations'].keys():    
        # appending relations
        for rel in response['annotations']['Relation']: 
            annotations.append({
                'type': 'Relation',
                '_subtype': rel['type'],
                'string': rel['string'],
                'start': rel['start'],
                'end': rel['end']
            })
        
    return annotations
```

```python
entity_list = ['Company','Person','Location','City','Product', 'Organization']
annotations = get_annotations(response, entity_list)
annotations = sorted(annotations, key=lambda k: k['start']) 
annotations
```

```python
from toolz import unique
from operator import itemgetter
print(len(annotations))
u_annotations = list(unique(annotations, key=itemgetter('string')))
len(u_annotations)
```

```python
set([a['_subtype'] for a in u_annotations])
```

```python
for entity in [a for a in u_annotations]:
    replacement = '<mark data-entity="{}">{}</mark>'.format(
        entity['_subtype'].lower(), entity['string'])
    body = body.replace(entity['string'],  replacement)
body
```

```python
html = """
<!DOCTYPE html>
<html>
<head>
  <link rel="stylesheet" href="mark.css">
</head>
<body>
<h1>The mark element</h1>
{}
</body>
</html>
""".format(body)
Html_file= open("output.html","w")
Html_file.write(html)
Html_file.close()
```
